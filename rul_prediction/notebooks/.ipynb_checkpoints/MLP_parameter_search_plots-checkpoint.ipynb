```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
INPUT_PATH = './experiment_set_14/parameter_search.csv'
OUTPUT_PATH = './experiment_set_14/plots'
```

```python
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)
```

```python
df = pd.read_csv(INPUT_PATH)
```

```python
df
```

```python
df['widest_layer'] = df['layers'].apply(lambda x: max([int(i) for i in x.split(' ')]))
```

```python
df['narrowest_layer'] = df['layers'].apply(lambda x: min([int(i) for i in x.split(' ')]))
```

```python
df
```

```python
df.sort_values(by=['loss_val_mean', 'loss_train_mean']).head(30)
```

```python
def create_boxplot(df, target_col, groupby_col, title="", x_label="", y_label="", output_path=None):
    labels = df[groupby_col].unique()
    labels = np.sort(labels)
    
    plot_data = []
    for label in labels:
        plot_data.append(df[df[groupby_col] == label][target_col])
        
    fig, ax = plt.subplots(figsize=(12, 7))
    ax.yaxis.set_ticks_position('none')
    ax.grid(color='grey', axis='y', linestyle='-', linewidth=0.25, alpha=0.5)

    ax.set_title(title, fontsize=20)
    ax.set_xlabel(x_label, fontsize=16)
    ax.set_ylabel(y_label, fontsize=16)

    ax.boxplot(plot_data, labels=labels, showmeans=True)
    
    if output_path is not None:
        plt.savefig(output_path, format='png', dpi=300)
```

```python
create_boxplot(df, "loss_val_mean", "num_layers", title="Effect of model depth", 
               x_label="Number of layers", y_label="Validation loss",
               output_path=os.path.join(OUTPUT_PATH, 'boxplot_layers.png'))
```

```python
create_boxplot(df, "loss_val_mean", "batch_size", title="Effect of batch size",
               x_label="Batch size", y_label="Validation loss",
               output_path=os.path.join(OUTPUT_PATH, 'boxplot_batch_size.png'))
```

```python
create_boxplot(df, "loss_val_mean", "activation", title="Effect of activation function",
               x_label="Activation", y_label="Validation loss",
               output_path=os.path.join(OUTPUT_PATH, 'boxplot_activation.png'))
```

```python
create_boxplot(df, "loss_val_mean", "dropout", title="Effect of dropout",
               x_label="Dropout probability", y_label="Validation loss",
               output_path=os.path.join(OUTPUT_PATH, 'boxplot_dropout.png'))
```

```python
create_boxplot(df, "loss_val_mean", "widest_layer", title="Effect of model width", 
               x_label="Number of nodes in widest layer", y_label="Validation loss",
               output_path=os.path.join(OUTPUT_PATH, 'boxplot_layer_width_max.png'))
```

```python
create_boxplot(df, "loss_val_mean", "narrowest_layer", title="Effect of model width", 
               x_label="Number of nodes in narrowest layer", y_label="Validation loss",
               output_path=os.path.join(OUTPUT_PATH, 'boxplot_layer_width_min.png'))
```

```python
def total_param(l=[]):
    s=0
    for i in range(len(l)-1):
        s=s+l[i]*l[i+1]+l[i+1]
    return s
```
